### Preparacion de datos para el entrenamiento

En esta parte separaremos los datos de entrenamiento, los escalaremos y transfomaremos.

```python
data_x["USD"].skew()
```

La funcion skew nos dice que tan asimetrico es la distribucion de las variables, como tenemos solo una variable continua usaremos la transformacion log en "USD" para simetrizar y acercarlo a una distribucion normal.

```python
Y=np.log(data_x["USD"])
Y.skew()
```

```python

X=data_x.drop(columns=["USD"],axis=1)
X=pd.DataFrame(MinMaxScaler().fit_transform(X),columns=X.columns)
```

##Modelos

```python
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import GridSearchCV

```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.3,random_state=0)
X_test,X_val,y_test,y_val=train_test_split(X_test,y_test,test_size=0.3,random_state=0)
```

###XGBRegressor

```python
gbR_grid = GradientBoostingRegressor()
gr_space = {
    'max_depth': [3,7,10],
    'n_estimators': [100, 300, 700, 1000],
    'max_features': [10, 20],
    'learning_rate': [0.1, 0.25, 0.05]
}

grid = GridSearchCV(gbR_grid, gr_space, cv = 2, scoring='r2', verbose = 3)
gbR_model_grid = grid.fit(X_train, y_train);

print('Best hyperparameters are '+str(gbR_model_grid.best_params_))
print('Best score is: ' + str(gbR_model_grid.best_score_))
```

```python
gbr = GradientBoostingRegressor(n_estimators=700,max_depth=10,max_features=20,learning_rate=0.1)
gbr.fit(X_train,y_train)
```

```python
print("score : ",gbr.score(X_test,y_test))
print("r2 : ",r2_score(y_test, gbr.predict(X_test)))
```

```python
plt.scatter(y_val, gbr.predict(X_val))
plt.legend(["r2 : %" + str(r2_score(y_val, gbr.predict(X_val)).round(4)*100)])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 5))
ax.hist(np.exp(y_val),bins=120,color="#c46464")
ax.hist(np.exp(gbr.predict(X_val)),bins=120,color="#86bf91")

ax.set_xlabel("USD")
ax.legend(["USD real","USD predecido"])
ax.set_title("Comparación de los dolares predecidos con los reales con el modelo GradientBoostingRegressor")
```

El metodo GradientBoostingRegressor dio buenos resultados con los parametros elegidos, se destaca una cierta incertidumbre en valores altos, pero una **alta** precisión en viviendas de relativamente bajo valor

###Ramdom forest

```python
rfr_grid = RandomForestRegressor()
gr_space = {
    'max_depth': [3,7,10],
    'n_estimators': [100, 200, 300, 400, 500],
    'max_features': [10, 20],
    'min_samples_leaf': [1, 2, 4]
}

grid = GridSearchCV(rfr_grid, gr_space, cv = 2, scoring='r2', verbose = 3);
rfr_model_grid = grid.fit(X_train, y_train);

print('Best hyperparameters are '+str(rfr_model_grid.best_params_))
print('Best score is: ' + str(rfr_model_grid.best_score_))
```

```python
rfr = RandomForestRegressor(n_estimators=300,max_depth=10,max_features=20,min_samples_leaf=1)
rfr.fit(X_train,y_train)
```

```python
print("score : ",rfr.score(X_test,y_test))
print("r2 : ",r2_score(y_test, rfr.predict(X_test)))
```
